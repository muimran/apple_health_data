```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import apple halth data
df = pd.read_csv("healthdata.csv", parse_dates=[1,2])
df.tail()
```

```python
#remove unnecessary columns
df = df.drop(labels= ["unit", "productType", "endDate"], axis= 1)
df.tail()
```

```python
#rename column
df = df.rename(columns={"startDate": "date", "value": "distance"})
```

```python
#convert date column to datetime
df['date'] = pd.to_datetime(df['date'])
df.tail()
```

```python
#change timezone of 'df'
df['date'] = df['date'].dt.tz_convert('Asia/Dhaka')
df.tail()
```

```python
#set 'date' column as index
df = df.set_index('date')
print(df)

#upsample dataframe to Day frequency
df= df.resample("D").sum()

#reset index
df = df.reset_index()
print(df)
df.head()
```

```python
#import dhaka temperature dataset
df_temp = pd.read_csv ("dhaka_temp.csv")
df_temp.head()
```

```python
#rename 'datetime' column to 'date'
df_temp = df_temp.rename(columns={"datetime": "date"})
df_temp.head()
```

```python
#convert date column to daetime
df_temp['date'] = pd.to_datetime(df_temp['date'])
df_temp.head()
```

```python
df_temp['date']=pd.to_datetime(df_temp["date"])

print(df_temp)
```

```python
#set 'date' column as index
df_temp = df_temp.set_index('date')
print(df_temp)
```

```python
#localize date column to Dhaka timezone
df_temp = df_temp.tz_localize(tz= 'Asia/Dhaka')
print(df_temp)
```

```python
#merge two dataframes
main_df = pd.merge(df,df_temp,on='date',how='left').fillna(0)
print(main_df)
```

```python
print(main_df)
main_df.head()

main_df = main_df.set_index('date')
main_df.tail()
```

```python
main_df= main_df.resample("M").sum()
main_df.tail()
```

```python
column_1 = main_df["distance"]
column_2 = df_temp["humidity"]

correlation = column_1.corr(column_2)
print(correlation)
```
